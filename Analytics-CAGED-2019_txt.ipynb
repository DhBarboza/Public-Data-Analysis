```python
identifica_sexo = {1:'Homem', 2:'Mulher', 9:'Não Identificado'}
identifica_situacao = {1:'Empregado', 2:'Desempregado', 9:'Não Identificado'}
identifica_deficiencia = {1:'Física',
                          0:'Não Deficiente',
                          2:'Auditiva',3:'Visual',
                          4:'Intelectual',
                          5:'Multipla',
                          6:'Reabilitado',
                          9:'Não Identificado'}
identifica_uf = {11:'Rondônia',
                12:'Acre',
                13:'Amazonas',
                14:'Roraima',
                15:'Pará',
                16:'Amapá',
                17:'Tocantins',
                21:'Maranhão',
                22:'Piaui',
                23:'Ceará',
                24:'Rio Grande do Norte',
                25:'Paraíba',
                26:'Pernambuco',
                27:'Alagoas',
                28:'Sergipe',
                29:'Bahia',
                31:'Minas Gerais',
                32:'Espírito Santo',
                33:'Rio de Janeiro',
                35:'São Paulo',
                41:'Paraná',
                42:'Santa Catarina',
                43:'Rio Grande do Sul',
                50:'Mato Grosso do Sul',
                51:'Moto Grosso',
                52:'Goiás',
                53:'Distrito Federal',
                99:'Não Identificado'}
identifica_estabelecimento = {1:'CNPJ',
                             3:'CAEPF',
                             4:'CNO',
                             5:'CEI',
                             9:'Não Identificado'}
```

```python
df_1.loc[(df_1['Sexo'].isnull()), 'Sexo'] = 9
df_1['Sexo'] = df_1['Sexo'].apply(lambda x: identifica_sexo[x])

df_1.loc[(df_1['Admitidos/Desligados'].isnull()), 'Admitidos/Desligados'] = 9
df_1['Admitidos/Desligados'] = df_1['Admitidos/Desligados'].apply(lambda x: identifica_situacao[x])

df_1.loc[(df_1['Tipo Defic'].isnull()), 'Tipo Defic'] = 9
df_1['Tipo Defic'] = df_1['Tipo Defic'].apply(lambda x: identifica_deficiencia[x])

df_1.loc[(df_1.UF.isnull()), 'UF'] = 9
df_1['UF'] = df_1['UF'].apply(lambda x: identifica_uf[x])

df_1.loc[(df_1['Tipo Estab'].isnull()), 'Tipo Estab'] = 9
df_1['Tipo Estab'] = df_1['Tipo Estab'].apply(lambda x: identifica_estabelecimento[x])
```
